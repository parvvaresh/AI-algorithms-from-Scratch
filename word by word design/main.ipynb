```python
import pandas as pd
```

```python
class word_by_word:
  def __init__(self):
    pass

  def fit(self, sents, k):
    self.k = k
    self.sents = list(map(lambda sent : sent.split(), sents))
    self.df = pd.DataFrame()
    words = list()
    for sent in self.sents:
      for word in sent:
        if word not in words:
          words.append(word)
    self.df["words"] = words
    self.df =  self.df.set_index("words")
    self._add_columns(words)
    self._fit_matrix()
    return self.df

  def _add_columns(self, words):
    for word in words:
      self.df[word] = None
  def _fit_matrix(self):
    for col_1 in self.df.columns:
      col_score = list()
      for col_2 in self.df.columns:
          col_score.append(self._counter_word(col_1, col_2))
      self.df[col_1] = col_score


  def _counter_word(self, word1, word2):
    count = 0
    for sent in self.sents:
      if word1 in sent and word2 in sent:
        if abs(sent.index(word1) - sent.index(word2)) <= self.k:
          count += 1
    return count
```

```python
model = word_by_word()
sents = [
    "ali reza parvaresh",
    "i am  ali reza and ai eng"
]
model.fit(sents, k = 2)
```
