```python
class DFS:
  def __init__(self):
    pass

  def dfs(self, graph, start):
    visited = set()
    stack = set([start])

    while stack:
      vartex = stack.pop()
      if vartex not in visited:
        visited.add(vartex)
        neighbers = graph[vartex]

        for neighber in neighbers:
          if neighber not in visited:
            stack.add(neighber)
    return visited
```

```python
model = DFS()
graph = {'0': set(['1', '2']),
         '1': set(['0', '3', '4']),
         '2': set(['0']),
         '3': set(['1']),
         '4': set(['1','2', '3'])}

model.dfs(graph, '0')
```
