```python
import string
import numpy as np
```

```python
class TER:
  def __init__(self):
    persian_punctuations = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ'''
    self.punctuations_list = string.punctuation + persian_punctuations

  def fit(self, ref, pred):
    ref_tokens = self._preprocess(ref)
    pred_tokens = self._preprocess(pred)
    self._matrix_edit_distance(ref_tokens, pred_tokens)

  def _matrix_edit_distance(self, ref_tokens, pred_tokens):
    cost = np.zeros((len(pred_tokens) + 1, len(ref_tokens) + 1))
    cost[0] = [j for j in range(0, len(ref_tokens) + 1)]
    cost[ : ,0] = [j for j in range(0, len(pred_tokens) + 1)]

    for row in range(1, len(pred_tokens) + 1):
      for col in range(1, len(ref_tokens) + 1):
        if ref_tokens[col - 1] == pred_tokens[row - 1]:
          cost[row][col] = cost[row - 1][col - 1]
        else:
          substitution = cost[row - 1][col - 1]
          insert = cost[row][col - 1]
          delete = cost[row - 1][col]
          cost[row][col] = min(insert, delete, substitution) + 1
    return cost




  def _preprocess(self, text):
    text = text.lower()
    translator = str.maketrans('', '', self.punctuations_list)
    text = text.translate(translator)
    return text.split() #return tokens
```

```python
model = TER()
model.fit("i am ali reza", "i am ali")
```
