```python
import pandas as pd
import numpy as np
```

```python
class Confusion_matrix:
  def __init__(self):
    pass
  def fit(self, y, y_predict):
    self.y = y
    self.y_predict = y_predict
    self.labels = list(set(y))
    self.df = pd.DataFrame()
    self.df["labels"] = self.labels

    for label_1 in self.labels:
      counter = list()
      for label_2 in self.labels:
        counter.append(self._counter(label_1, label_2))
      self.df[label_1] = counter

    self.df = self.df.set_index("labels")

  def _counter(self, label_1, label_2):
    counter = 0
    for index in range(0, len(self.y)): #or len of self.y_predict
      if self.y[index] == label_1 and self.y_predict[index] == label_2:
        counter += 1
    return counter
  def get_df(self):
    return self.df

```

```python
y = np.array([1 , 1, 0, 0, 1])
y_pred = np.array([0 , 1, 0, 1, 1])
```

```python
model = Confusion_matrix()
model.fit(y, y_pred)
model.get_df()
```
